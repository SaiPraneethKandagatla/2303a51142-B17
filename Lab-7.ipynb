```
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("world_population.csv")
plt.figure(figsize=(10, 6))
plt.plot(df["Year"], df["Population"], marker="o", color="b", linewidth=2)
plt.xlabel("Year")
plt.ylabel("World Population (Billions)")
plt.title("World Population Growth Over Years")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("AAPL.csv", parse_dates=["Date"])
plt.figure(figsize=(12, 6))
plt.plot(df["Date"], df["Close"], marker="o",color="g", linewidth=2)
plt.xlabel("Date")
plt.ylabel("Closing Price (USD)")
plt.title("Apple Stock Price Trend")
plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```
